MLP
- Splitting dataset parsing by context length (, 3) => 3 is length of context window
- nn layers:
- input layer -  (6 [a.k.a (3, 2) flattened], 100 )
- hidden layer - (100, 100)
- non-linear layer - tanh
- hidden layer - (100, 100)
- non-linear layer - softmax
- output layer - (100, 27???)

```python
# encoding/decoding functions

chrs = sorted(list(set(string)))

stoi = {chr: i for i, chr in enumerate(chrs[2:])}
stoi["."] = 26
itos = {i: chr for chr, i in stoi.items()}

def encode(str):
  return [stoi[chr] for chr in str]

def decode(nums):
  return [itos[int(num)] for num in str(nums)]
```

```python
import torch
import torch.nn.functional as F

# read from dataset then split into names
word_dataset = string.splitlines()

# split into block_size a.k.a context size
# block_size = 3

X = []
y = []

i = 0
for word in word_dataset[:100]:
  word = ["."] + list(word) + ["."]
  for i in range(len(word) - 3):
    comb = word[i] + word[i + 1] + word[i + 2]
    res = word[i + 3]
    comb = encode(comb)
    res = encode(res)

    X.append(comb)
    y.append(res)

Xs = torch.tensor(X, dtype=torch.long)
# ys = torch.tensor(y, dtype=torch.float16)

Xs.shape
```

```python
# embedding matrix

embedding = torch.rand((27, 2), dtype=torch.float16)

# neural network layers
h1 = torch.rand((6, 8), dtype=torch.float16)
h2 = torch.rand((8, 27), dtype=torch.float16)
```

```python
onehot = F.one_hot(Xs[0], 27)
onehot = onehot.to(torch.float16)
embedded_matrix = onehot @ embedding
input = embedded_matrix.view(-1, 6)
input.shape
```

```python
# foward pass
out1 = input @ h1
nonlin1 = F.tanh(out1)
out2 = nonlin1 @ h2
# outputs = F.softmax(out2, dim=0)
# -outputs[y[0]].log()
out2.shape
```

```python
# calculate loss

loss = F.cross_entropy(out2, torch.tensor(y[0], dtype=torch.long).view(-1))
loss
# torch.tensor(y, dtype=torch.long).view(-1).shape
```
